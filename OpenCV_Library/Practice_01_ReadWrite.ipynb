```python
while(vid.isOpened()):
# Capture frame-by-frame
    ret, frame = vid.read()
    if ret == True:
# Display the resulting frame
        cv2.imshow('Frame',frame)
        if cv2.waitKey(25) & 0xFF == ord('d'):
            break
# Break the loop
    else: 
        break
# When everything done, release the video capture object
vid.release()
 
# Closes all the frames
cv2.destroyAllWindows()
```

```python
# Function to resize an image
def rescaleFrame(frame, scale=0.75):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    dimensions = (width,height)
    return cv2.resize(frame, dimensions, interpolation=cv2.INTER_AREA)
```

```python
resize_pic = rescaleFrame(img)
cv2.imshow("test", resize_pic)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
while(vid.isOpened()):
# Capture frame-by-frame
    ret, frame = vid.read()
    if ret == True:
# Display the resulting frame
        frame_res = rescaleFrame(frame)
        cv2.imshow('Frame',frame_res)
        if cv2.waitKey(25) & 0xFF == ord('d'):
            break
# Break the loop
    else: 
        break
# When everything done, release the video capture object
vid.release()
 
# Closes all the frames
cv2.destroyAllWindows()
```

# DRAWING using OpenCV

```python
# Create a blank image
blank = np.zeros((500,500,3), dtype='uint8')
cv2.imshow("Blank Image",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def create():
    return np.zeros((500,500,3), dtype='uint8')
```

```python
# Paint the blank image a color
blank = create()
blank[:] = 0,0,255

cv2.imshow("Red",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Color a specific part
blank = create()
blank[200:300 , 300:400] = 0,0,255

cv2.imshow("Red",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rectangle using opencv
blank = create()
cv2.rectangle(blank,(0,0),(250,500),(0,255,0),thickness=-1)
cv2.imshow("Rectagle",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Cirle using opencv2
blank = create()
cv2.circle(blank,(250,250),80,(255,0,0),thickness=10)
cv2.imshow("Circle",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Lines usig opencv2
blank = create()
cv2.line(blank,(100,250), (300,400), (255,255,255), thickness=1)
cv2.imshow("Line",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
 # Text using opencv2
blank = create()
cv2.putText(blank,"Hello World!",(255,255),cv2.FONT_HERSHEY_TRIPLEX, 1.0, (0,255,0),thickness=2)
cv2.imshow("Text",blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
